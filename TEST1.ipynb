```python
import math


def arrangement(n, k):
    A = math.factorial(n)/math.factorial(n - k)
    return A
```

```python


def combination(n, k):
    C = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    return C
```

```python
from matplotlib import pyplot as plt
import numpy as np

def plane_time(t_wait):
    interval = 6
    time_per_path = 1/2
    return time_per_path + interval + t_wait

def bus_time(t_wait):
    interval = 6
    time_per_path = 1/2
    return time_per_path + interval + t_wait

time_max = 12

x = np.arange(0,time_max)
y = []

for t in range(time_max):
    y.append(plane_time(t))

fig, ax = plt.subplots()

ax.plot(x, y)

y = []

for t in range(time_max):
    y.append(bus_time(t))

ax.plot(x, y)
```
