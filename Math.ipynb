```python
import math


def arrangement(n, k):
    A = math.factorial(n)/math.factorial(n - k)
    return A
```

```python


def combination(n, k):
    C = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    return C
```

```python
class Box:
    def __init__(self, n_white, n_black):
        self.n_white = n_white
        self.n_black = n_black
    
    @property
    def n(self):
       return self.n_white + self.n_black


box1 = Box(3, 1)
box2 = Box(6, 4)
box3 = Box(4, 1)

p_white_box1 = box1.n_white / box1.n
p_white_box2 = box2.n_white / box2.n
p_white_box3 = box3.n_white / box3.n

p = p_white_box1 * p_white_box2 * p_white_box3 

p
```
