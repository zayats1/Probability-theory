# Статистика

## Основні поняття математичної статистики:

• Групу об’єктів, поєднаних за якоюсь якісною
чи кількісною ознакою, називають
**статистичною сукупністю**.

• Сукупність усіх об’єктів, що підлягають
дослідженню називають **генеральною**.

• Якщо із генеральної сукупності вибрати
тільки деякі елементи випадковим чином, то
одержимо вибіркову сукупність, яку
називають **вибіркою**.

### Вибірка



1. Випадкова   
    Вибір здійснюється жеребкуванням  



2. Механічна
    Сукупність розбивається на групи і від кожної вибирають представника  

3. Типова
    Вибір у певному порядку (наприклад,кожен десятий)  

### Для наочного зображення інформації про вибірку у статистиці використовують:

Стовпчасті діаграми зі з'єднаних прямокутників, які
називають **гістограмами**.

<img src ="https://upload.wikimedia.org/wikipedia/commons/thumb/8/8e/Histogram_example.svg/1200px-Histogram_example.svg.png" width = 500 >

Якщо з'єднати відрізками середини верхніх основ послідовних прямокутників гістограми, то побудуємо **полігон частот**.

Інші види діаграм (кругова, полярна) та графіки.

• Всі зібрані дані заносять в таблиці, які називаються 
**статистичними**.  
• У математичній статистиці найпоширенішими є
**частотні** та **інтервальні таблиці**.

### Об'єм вибірки

Сума всіх частот досліджуваної вибірки називається **об'ємом цієї вибірки**: n

```python
sample_size = len(data)
sample_size
```

### Медіана

Медіана – це середня величина змінюваної ознаки, яка
міститься всередині варіаційного ряду  
 (число, яке ділить
варіаційний ряд навпіл або середнє арифметичне двох
чисел)

```python
import statistics as stat
the_median = stat.median(data)
the_median
```

### Мода

Мода – це значення варіанти, яка трапляється найчастіше

```python
stat.mode(data)
```

### Середнє арифметичне



Середнє арифметичне значення знаходять за
формулою: x =

(x1+x2+⋯+xn) / n

```python
arithmetic_mean = stat.mean(data)
arithmetic_mean
```

### Дисперсія

Дисперсія, або другий момент відносно середнього, є мірою мінливості (розкиду або поширення) даних.

```python
stat.variance(data)
```

### Відхилення 

Відхилення знаходять за формулою: (xi − ar_mean(x))

```python
def dev(xi, ar_mean): return xi - ar_mean
```

### Середнє квадратичне відхилення

```python
stat.stdev(data)
```

> якщо середньє арифметичне вирахуване, то, щоб заново його не рахувати, можна передати його у параметр цієї функції

```python
stat.stdev(data,arithmetic_mean)
```

### Частоти

Кількість елементів кожної варіанти називають її **частотою**:  
> **Наприклад:** вибірка елементів зі значенням 11 містить 5 штук

```python
import pandas as pd
frequencies = pd.array(data).value_counts()
frequencies
```

Відношення частоти значення до об'єму вибірки називається **відносною частотою** : ωi = ni/n

```python
def relative_frequency(nI, n): return ni/n
```

### Середнє геометричне

```python
stat.geometric_mean(data)
```
