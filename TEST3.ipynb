```python
import math


def arrangement(n, k):
    A = math.factorial(n)/math.factorial(n - k)
    return A


def combination(n, k):
    C = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    return C


def bernouli_formula(p, n, k):
    q = 1 - p
    P = combination(n, k) * pow(p, k)*pow(q, n-k)
    return P


def poisson(n, p, k):
    jl = p*n
    P = (pow(jl, k) / math.factorial(k)) * math.exp(-jl)
    return P

def binomial_distrtibution(p, n):
    distribution = list({})
    for k in range(n + 1):
        p_k = bernouli_formula(p, n, k)
        distribution.append(p_k)

    return distribution
    
def poisson_distrtibution(p, n):
    distribution = list({})
    for k in range(n + 1):
        p_k = poisson(n, p, k)
        distribution.append(p_k)

    return distribution

def print_distribution(distribution,precession):
    sum = 0
    for k, p_k in enumerate(distribution):
        sum += p_k
        print(f"p{k} = {p_k:.precession}")
    print(f"sum = {sum:.4}")
```
